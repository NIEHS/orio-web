In hg19, there are duplicate entries for a select few older files.  Newer entries may be distinguished by presence of a dateResubmitted field.

```python
def readMetaDataIntoList(meta_file):
    f = open(meta_file)
    meta_dict = dict()
    for line in f:
        file_name = line.split("\t")[0]
        if file_name not in meta_dict:
            meta_dict[file_name] = []
        entry_dict = dict()
        for entry in line.split("\t")[1].split(";"):
            entry_dict[entry.strip().split("=")[0]] = entry.strip().split("=")[1]
        meta_dict[file_name].append(entry_dict)
    return meta_dict
```

```python
def checkUniqueFileNames(meta_dict):
    for file_name in meta_dict:
        if len(meta_dict[file_name]) > 1:
            return False
    return True
```

```python
def readMetaData(meta_file):
    f = open(meta_file)
    meta_dict = dict()
    for line in f:
        file_name = line.split("\t")[0]
        #if file_name not in meta_dict:
        #    meta_dict[file_name] = []
        entry_dict = dict()
        for entry in line.split("\t")[1].split(";"):
            entry_dict[entry.strip().split("=")[0]] = entry.strip().split("=")[1]
        #if "md5sum" in entry_dict:
        #    meta_dict[file_name] = entry_dict
        if file_name in meta_dict:
            if "dateResubmitted" in entry_dict:
                meta_dict[file_name] = entry_dict
        else:
            meta_dict[file_name] = entry_dict
    return meta_dict
```

```python
def getKeys(meta_dict):
    key_list = []
    for file_name in meta_dict:
        for key in meta_dict[file_name]:
            if key not in key_list:
                key_list.append(key)
    return key_list
```

Associate stranded bigwigs together

```python
different_keys = []
for file_name in hg19_meta:
    if "Plus" in file_name:
        minus_name = file_name.replace("Plus","Minus")
        for entry in hg19_meta[file_name]:
            if entry in hg19_meta[minus_name]:
                if hg19_meta[file_name][entry] != hg19_meta[minus_name][entry]:
                    if entry not in different_keys:
                        different_keys.append(entry)
            else:
                if entry not in different_keys:
                        different_keys.append(entry)
print(different_keys)
```

```python
def associateStranded(meta_dict):
    stranded_fields = ['tableName', 'view', 'md5sum', 'size']
    associated_dict = dict()
    for file_name in meta_dict:
        if "Plus" in file_name:
            entry_name = file_name.replace(".bigWig","").replace("Plus","")
            if entry_name not in associated_dict:
                associated_dict[entry_name] = dict()
            associated_dict[entry_name]["plus_bigwig"] = file_name
            for field in stranded_fields:
                if field in meta_dict[file_name]:
                    associated_dict[entry_name]['plus_' + field] = meta_dict[file_name][field]
            for field in meta_dict[file_name]:
                if field not in stranded_fields:
                    associated_dict[entry_name][field] = meta_dict[file_name][field]
        elif "Minus" in file_name:
            entry_name = file_name.replace(".bigWig","").replace("Minus","")
            if entry_name not in associated_dict:
                associated_dict[entry_name] = dict()
            associated_dict[entry_name]["minus_bigwig"] = file_name
            for field in stranded_fields:
                if field in meta_dict[file_name]:
                    associated_dict[entry_name]['minus_' + field] = meta_dict[file_name][field]
            for field in meta_dict[file_name]:
                if field not in stranded_fields:
                    associated_dict[entry_name][field] = meta_dict[file_name][field]
        else:
            entry_name = file_name.replace(".bigWig","")
            associated_dict[entry_name] = dict()
            associated_dict[entry_name]["ambig_bigwig"] = file_name
            for field in stranded_fields:
                if field in meta_dict[file_name]:
                    associated_dict[entry_name]['ambig_' + field] = meta_dict[file_name][field]
            for field in meta_dict[file_name]:
                if field not in stranded_fields:
                    associated_dict[entry_name][field] = meta_dict[file_name][field]
    return associated_dict

hg19_final = associateStranded(hg19_meta) 
mm9_final = associateStranded(mm9_meta)
```

```python
hg19_keys = getKeys(hg19_final)
mm9_keys = getKeys(mm9_final)

all_keys = hg19_keys + list(set(mm9_keys) - set(hg19_keys))
```

```python
print(all_keys)
```

```python
OUTPUT = open("combined_ENCODE_metadata.txt", "w")

for key in all_keys:
    OUTPUT.write("\t" + key)
OUTPUT.write("\n")

def writeMetaData(meta_data, OUTPUT):
    for entry in meta_data:
        OUTPUT.write(entry)
        for key in all_keys:
            if key not in meta_data[entry]:
                OUTPUT.write("\t")
            else:
                OUTPUT.write("\t" + meta_data[entry][key])
        OUTPUT.write("\n")

writeMetaData(hg19_final, OUTPUT)
writeMetaData(mm9_final, OUTPUT)

OUTPUT.close()
```
